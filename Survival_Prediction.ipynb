```python
dataset = pd.read_csv('/content/drive/My Drive/Dataset/train.csv')
print(dataset)
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
X = dataset.iloc[:,[2,4,5,6,7,9]].values
y = dataset.iloc[:, 1].values
```

```python
X
```

```python
y
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.NAN, strategy= 'median', fill_value='None', verbose=1, copy=True)
imputer  = imputer.fit(X[:, 2:3])
X[:, 2:3] = imputer.transform(X[:, 2:3])
X
```

```python
pd.DataFrame(X, columns = ['Pclass',' Sex', 'Age', 'SibSp', 'Parch', 'Fare']).isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
X[:, 1] = labelencoder_X.fit_transform(X[:, 1])
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train 
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier =  KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)  #minkowski=1(manhaton), minkowski=2(eculidian) 
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix: ')
print(cm)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %.2f' % (accuracy_score(y_test, y_pred)*100))
```

```python
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred, average='binary')
print('Precision: %.3f' % (precision*100))
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
recall = recall_score(y_test, y_pred, average='binary')
print('Recall: %.3f' % (recall*100))
```

```python
score = f1_score(y_test, y_pred, average='binary')
print('F-Measure: %.3f' % (score*100))
```

```python
import joblib
filename = '/content/drive/My Drive/Survival_Prediction.sav'
joblib.dump(classifier, filename)
loaded_model = joblib.load(filename)
result = loaded_model.score(X_test, y_test)
print(result)
```
